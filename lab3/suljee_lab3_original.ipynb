## Граф бүтэц, түүний дүрслэл

``networkx`` нь хэд хэдэн граф моделийг дэмждэг. Үүнд

*   undirected графуудыг үүсгэхдээ ``Graph`` класс
*   directed графуудыг дүрслэхдээ ``DiGraph`` класс

Энэ удаа бид зөвхөн undirected буюу чиглэлгүй графуудыг авч үзнэ.

We can build a graph adding nodes as well as edges as follows:

```python
g = nx.Graph()

g.add_node("a")
g.add_edge("a", "b")
g.add_edge("a", "c")
g.add_edge("b", "c")

nx.draw(g ,with_labels=True)
```

Графын орой болон ирмэгүүдийг устгаж бас болно.

```python
g.remove_node("a")
g.remove_edge("b", "c")

#re-draw the graph
nx.draw(g, with_labels=True)
```

## Оройн зэрэг ба зэргийн тархалт
Тухайн өгөгдсөн оройн зэргийг нь дараах кодоор амархан авч болно. Жишээлбэл, энд 1 гэсэн дугаартай оройн зэргийг олж байна.

Үүний адилаар дундаж оройн зэргийг бодож болно.

Оройн зэргийн тархалтыг дараах аргаар хялбархан бодож, графикаар харуулж болдог.

```python
hist = nx.degree_histogram(g)
    
plt.plot(range(0, len(hist)), hist, ".")
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("#Nodes")
plt.loglog()
plt.show()
```

Кодоо илүү сайжруулбал ...

```python
import scipy as sp
```

```python
def plot_dist(G):
    M = nx.to_scipy_sparse_matrix(G)
    xmin = min([d[1] for d in G.degree()])
    indegrees = M.sum(0).A[0]
    degree = np.bincount(indegrees)
    fit = powerlaw.Fit(np.array(degree)+1, fit_method='KS')#, xmin=xmin, xmax=max(degree)-xmin,discrete=True)
    
    
    fig = plt.figure(figsize=(16, 6)) 
    """ Plot Distribution """
    plt.subplot(1, 3, 1)
    plt.plot(range(len(degree)),degree,'b.')   
    plt.loglog()
    plt.xlim((min(degree), max(degree)))
    plt.xlabel('Degree')
    plt.ylabel('P(k)')


    """ Plot CDF """
    plt.subplot(1, 3, 2)
    fit.plot_cdf()
    plt.xlabel("Degree")
    plt.ylabel('CDF')

    """ Plot CCDF """
    plt.subplot(1, 3, 3)
    fit.plot_ccdf()
    plt.ylabel('CCDF')
    plt.xlabel('Degree')
    plt.tight_layout()
    plt.show()
```

```python
plot_dist(g)
```

## Граф компонентууд
``networkx`` нь үндсэн графын тодорхой харагдацуудыг сонгогдсон оройгоос харах боломжийг олгоно.

```python
list(g.neighbors(1)) # obtain the list of neighobors for node 0
```

```python
ego = nx.ego_graph(g, 1) # ego network of the node 0
nx.draw(ego, with_labels=True)
```

Графын хувьд бүх тусгаар холбогдсон хэсгүүд буюу connected component-уудыг задлаж харбал

```python
nx.number_connected_components(g)
```

```python
comps = list(nx.connected_components(g)) # get a list of connected components (for decreasing size)
comp_1 = nx.subgraph(g, comps[1]) # build a subgraph on the second component
nx.draw(comp_1)
```

## Triangles, density and clustering
Бусад сүлжээний метрик үзүүлэлтүүд ч уг санг ашигласнаар хялбар бодогдоно.

```python
nx.density(g)
```

```python
nx.triangles(g)[1] # count the triangles each node is involved in (and access the value of node 0)
```

```python
nx.clustering(g)[1] # compute the local clustering coefficient for all nodes (and access the value for node 0)
```

```python
nx.average_clustering(g) # compute the global clustering coefficient
```
